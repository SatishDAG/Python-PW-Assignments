## Question 1 — What is Dimensionality Reduction? Why is it important in machine learning?

**Answer:** Dimensionality reduction refers to techniques that reduce the number of input variables (features) while preserving the significant structure and information in the data. It is important because it: 

- Reduces computational cost and memory usage.
- Mitigates the curse of dimensionality and helps avoid overfitting.
- Removes noise and redundant features.
- Enables visualization and easier interpretation.


## Question 2 — Name and briefly describe three common dimensionality reduction techniques.

**Answer:**

1. **Principal Component Analysis (PCA):** A linear method that finds orthogonal directions (principal components) that capture maximum variance. Good for decorrelation and compression.

2. **t-SNE (t-Distributed Stochastic Neighbor Embedding):** A nonlinear technique tailored for visualization. It preserves local structure and reveals clusters in 2D/3D plots but is not ideal for downstream modeling.

3. **Linear Discriminant Analysis (LDA):** A supervised method that projects data to maximize class separability; useful when labels exist and dimensionality reduction for classification is required.


## Question 3 — What is clustering in unsupervised learning? Mention three popular clustering algorithms.

**Answer:** Clustering is the process of grouping unlabeled data points into clusters so that items in the same cluster are more similar to each other than to those in other clusters.

**Popular algorithms:** K-Means, DBSCAN, Hierarchical Agglomerative Clustering.


## Question 4 — Explain the concept of anomaly detection and its significance.

**Answer:** Anomaly detection identifies observations that significantly deviate from the expected pattern.

**Significance:**

- Fraud detection (finance)
- Intrusion detection (networks)
- Fault detection in industrial systems
- Medical outlier detection (rare diseases)


## Question 5 — List and briefly describe three types of anomaly detection techniques.

**Answer:**

1. **Statistical Methods:** Use parametric or nonparametric statistical tests (e.g., z-score, Grubbs') to identify unlikely observations.

2. **Distance-/Density-Based Methods:** Points with low local density or large distance to neighbors are flagged (e.g., LOF, DBSCAN).

3. **Machine-Learning Methods:** Include supervised (if labeled anomalies available), semi-supervised (one-class SVM, autoencoders), and ensemble methods (Isolation Forest).


## Question 6 — What is time series analysis? Mention two key components of time series data.

**Answer:** Time series analysis studies data points collected or indexed in time order to understand underlying structure and to forecast future values.

**Two key components:** Trend (long-term direction) and Seasonality (regular periodic fluctuations).


## Question 7 — Describe the difference between seasonality and cyclic behavior in time series.

**Answer:**

- **Seasonality:** Regular, fixed-period patterns (e.g., monthly, quarterly).

- **Cyclic behavior:** Irregular, longer-term oscillations often tied to economic or business cycles without a fixed period.


## Question 8 — Write Python code to perform K-means clustering on a sample dataset (with output).

**Answer (Code):** The following code generates synthetic data, runs K-Means, and plots clusters and inertia vs K.


```python
# Q8: K-Means clustering example
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data
X, _ = make_blobs(n_samples=400, centers=4, cluster_std=1.0, random_state=42)

# Plot data
plt.figure(figsize=(6,4))
plt.scatter(X[:,0], X[:,1], s=20, alpha=0.7)
plt.title('Synthetic data for clustering')
plt.show()

# Run KMeans and show cluster centers for a chosen K
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(X)

plt.figure(figsize=(6,4))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='tab10', s=20, alpha=0.8)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', s=150, marker='X')
plt.title(f'KMeans Clustering (K={k})')
plt.show()

# Elbow plot: inertia vs K
inertias = []
Ks = range(1,11)
for K in Ks:
    km = KMeans(n_clusters=K, random_state=42)
    km.fit(X)
    inertias.append(km.inertia_)

plt.figure(figsize=(6,4))
plt.plot(list(Ks), inertias, marker='o')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('Elbow plot for KMeans')
plt.grid(True)
plt.show()
```

## Question 9 — What is inheritance in OOP? Provide a simple example in Python.

**Answer:** Inheritance lets a class (child) reuse attributes and methods from another class (parent). Example below.


```python
# Q9: OOP Inheritance example
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return '...'

class Dog(Animal):
    def speak(self):
        return 'Bark'

# Usage
d = Dog('Rex')
print(d.name, 'says', d.speak())
```
